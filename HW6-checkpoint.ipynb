##First fitting a line using m and b formulas:

```python
%matplotlib inline  
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

x = np.array([300.,330,360,390,420,450,480,510,540,570,600])
y = np.array([47,40.1,36.7,33.7,31.8,30.3,29.3,28.6,28.1,27.7,27.4])

n = len(x)
m = (n*sum(xy)-sum(x)*sum(y))/(n*sum(x*x)-sum(x)**2)
b = (sum(x*x)*sum(y)-sum(x*y)*sum(x))/(n*sum(x*x)-sum(x)**2)
x_fit = np.linspace(min(x),max(x))
y_fit = m*x_fit+b

fig,ax = plt.subplots()
ax.scatter(x,y)
ax.plot(x_fit,y_fit,color='r')
ax.set_xlabel('Time in secs')
ax.set_ylabel('Temperature C')
title = 'The best fit line as a slope m={:.3f} and intercept b={:.3f}'.format(m,b)
ax.set_title(title)
```
